```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xgboost as xgb
```

```python
data = pd.read_csv('./data/data_4hr.csv')
data.head()
```

```python
data.Date = data.Date.apply(lambda x: pd.to_datetime(x))
data.set_index('Date', inplace=True)

data.head()
```

```python
data = data.reset_index()
```

```python
data.head()
```

Setting class labels to the dataframe. Will attempt classification model.

```python
data['Label'] = 0

for i in range(len(data.index)):
    data['Label'][i] = 0 if i == 0 else (1 if data['Close'][i] > data['Close'][i-1] else 0)
```

```python
data.tail()
```

```python
data = data.set_index('Date')

train = data[:6000]
test = data[6000:]

X_train = train.drop(['Close','Label'], axis=1)
X_test = test.drop(['Close', 'Label'], axis=1)
y_train = train.Label
y_test = test.Label
```

```python
X_train.dtypes
```

```python
model = xgb.XGBClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay


accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

confusion_matrix = confusion_matrix(y_test, y_pred)
```

```python
print(f'The accuracy score of the model is: {accuracy}')
print(f'The precision score of the model is: {precision}')
print(f'The recall score of the model is: {recall}')
print(f'The f1 score of the model is: {f1}')
```

```python
# plotting confusion matrix to get an idea of how model performed

display = ConfusionMatrixDisplay(confusion_matrix, display_labels = model.classes_)

display.plot()
plt.show()
```

The accuracy score of the model shows its ability to correctly classify the appropriate label. Given that this is intended for trading, anything over 50% should be deemed successful. However would ideally like to get this up above 60% if possible.

The recall score of the model shows its ability to correctly identify positive samples (in this case a positive sample indicates upward rpice movement). Having a high recall score can be quite advantageous even without a higher accuracy, being able to identify upward price movement nearly two-thirds of the time is a good result.
