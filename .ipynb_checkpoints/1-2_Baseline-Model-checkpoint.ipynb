```python
# import necessary packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.linear_model import Ridge
```

```python
# import csv from prior notebook into new dataframe

data = pd.read_csv('./data/data_4hr.csv')
```

```python
data = data.set_index('Date')
data.tail()
```

```python
data.shape
```

```python
# seperate into train and test, and then X and y

train = data[:6000]
test = data[6000:]
```

```python
train.shape
```

```python
test.shape
```

```python
X_train = train.drop('Close', axis=1)
X_test = test.drop('Close', axis=1)
y_train = train.Close
y_test = test.Close
```

```python
# initialize the baseline model

baseline = Ridge()

baseline.fit(X_train, y_train)
```

```python
# predict test variables

y_pred = baseline.predict(X_test)
```

```python
# evaluate baseline model

baseline_r2 = r2_score(y_test, y_pred)
baseline_MSE = mean_squared_error(y_test, y_pred)

print(f'Baseline model r2: {baseline_r2}\nBaseline MSE: {baseline_MSE}')
```

The baseline model has a very large mean squared error, none of the features have been scaled at this point, will attempt scaling of columns to see if it improves metrics.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
X_train_scaled
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, index=X_train.index, columns=X_train.columns)
```

```python
X_test_scaled = pd.DataFrame(X_test_scaled, index=X_test.index, columns=X_test.columns)
```

```python
X_train_scaled
```

```python
baseline.fit(X_train_scaled, y_train)
```

```python
y_pred = baseline.predict(X_test_scaled)
```

```python
# evaluating baseline model with scaled data

baseline_r2_1 = r2_score(y_test, y_pred)
baseline_MSE_1 = mean_squared_error(y_test, y_pred)

print(f'Baseline model r2: {baseline_r2_1}\nBaseline MSE: {baseline_MSE_1}')
```

Scaling the data in this case produced a much larger mean squared error and a much worse r2 score also.
